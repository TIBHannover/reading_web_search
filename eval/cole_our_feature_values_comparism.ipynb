```python
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
our_features = pd.read_csv("../out/our_content_page_features.csv")
cole2011_features = pd.read_csv("../out/cole2011_content_page_features.csv")
```

```python
df = pd.DataFrame()
df["our_median"] = our_features.median()
df["our_mean"] = our_features.mean()
df["our_std"] = our_features.std()
df["cole2011_median"] = cole2011_features.median()
df["cole2011_mean"] = cole2011_features.mean()
df["cole2011_std"] = cole2011_features.std()
df.to_markdown("../results/cole2011_vs_our_feature_values.md", index=False)
df
```
